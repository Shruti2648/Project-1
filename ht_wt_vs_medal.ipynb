```python
# Import dependencies:

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Read csv file and store into dataframe:

complete_df = pd.read_csv("../Resources/athlete_events.csv")
#complete_df
```

complete_df

```python
#cleanup by ht_wt .. there is no cleaner way 

cleaned_up_ht_wt = complete_df.dropna(subset= ["Height", "Weight",  "Medal"] )
#cleaned_up_ht_wt
```

```python
my_ht_cut_ratios = [ 0, 150 ,165, 180,195, 240 ]
my_ht_bin = (cleaned_up_ht_wt, my_ht_cut_ratios)
#my_ht_bin
```

```python
ht_medal  = cleaned_up_ht_wt [["Height" , "Medal"]]
#ht_medal
```

```python
bmi_medal = cleaned_up_ht_wt[["Height", "Weight", "Medal", "Sport"]]
print(bmi_medal.Sport.unique())
bmi_medal["BMI"] = bmi_medal["Weight"] / (bmi_medal["Height"] * bmi_medal["Height"])*10000
bmi_medal_rating = bmi_medal[["BMI","Medal"]]
bmi_medal_rating
wt_medal_count  = bmi_medal_rating.groupby(['BMI']).count()
wt_medal_count.plot(title="All  Medals by BMI",kind="line").get_figure().savefig('bmi_medal.png')
```

```python
#see theres a correaltion between sport type and BMI 
# find 5 sports and use those for the analysis 
# 1. Wrestling 
my_list = []

weight_bmi = bmi_medal.loc[bmi_medal['Sport'] == 'Wrestling']
weight_bmi["BMI"] = weight_bmi["Weight"] / (weight_bmi["Height"] * weight_bmi["Height"])*10000
my_list.append(weight_bmi.loc[weight_bmi['Sport'] == 'Wrestling'])
weight_bmi_1 = weight_bmi[["BMI","Medal"]]
wt_medal_count  = weight_bmi_1.groupby(['BMI']).count()
import copy
mine_medal = copy.deepcopy(wt_medal_count)
#ax = wt_medal_count.plot()
wt_medal_count.plot(title="Wrestling BMI",kind="line").get_figure().savefig('wrestling_bmi.png')
#my_list.append(ax)
```

```python
# 2. Swimming

weight_bmi = bmi_medal.loc[bmi_medal['Sport'] == 'Swimming']
weight_bmi["BMI"] = weight_bmi["Weight"] / (weight_bmi["Height"] * weight_bmi["Height"])*10000
weight_bmi_1 = weight_bmi[["BMI","Medal"]]
my_list.append(weight_bmi.loc[weight_bmi['Sport'] == 'Swimmming'])
wt_medal_count  = weight_bmi_1.groupby(['BMI']).count()
mine_medal_1 = copy.deepcopy(wt_medal_count)
wt_medal_count.plot(title="Swimming BMI",kind="line").get_figure().savefig('swimming.png')

#wt_medal_count.plot()
#my_list.append(ax)
```

```python
# Weight lifting
weight_bmi = bmi_medal.loc[bmi_medal['Sport'] == 'Weightlifting']
weight_bmi["BMI"] = weight_bmi["Weight"] / (weight_bmi["Height"] * weight_bmi["Height"])*10000
weight_bmi_1 = weight_bmi[["BMI","Medal"]]
my_list.append(weight_bmi.loc[weight_bmi['Sport'] == 'Weightlifting'])
wt_medal_count  = weight_bmi_1.groupby(['BMI']).count()
mine_medal_2 = copy.deepcopy(wt_medal_count)
wt_medal_count.plot(title="Weight Lifting BMI",kind="line")
#my_list.append(ax)

```
